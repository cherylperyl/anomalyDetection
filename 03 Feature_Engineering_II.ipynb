## Reaction Time

Reaction time is the time difference between two bidder's bids in the same auction on the same day.  
  
First retrieve the individual reaction time of each transaction record in `bids_df`.

```python
# shortcut to obtain reaction_time_raw_df
reaction_time_raw_df = pd.read_csv('raw_data/reaction_time_raw.csv')
```

```python
# create a starting dataframe
reaction_time_df = pd.DataFrame(columns=['bidder_id','time'])

# create list to iterate through
days = [day1,day2,day3,day4,day5,day6,day7,day8,day9]
current_day = 0 # to track output

for day in days:
    current_day += 1
    
    # get list of unique auction for the specific day
    unique_auctions = list(day.auction.unique())
    
    # iterate through the individual auctions
    for auction in unique_auctions:
        
        # get the transactions for that unique auction
        auction_df = day.loc[day.auction == auction].set_index('bidder_id').drop(columns='auction',axis=1)
        
        # get time diff
        auction_df = auction_df.diff()
        
        # reset index to append to reation_time_df
        auction_df = auction_df.reset_index()
        reaction_time_df = pd.concat([reaction_time_df, auction_df],ignore_index=True)
        
    # print(f'| Day {current_day} done |', end=' ') # to track output
```

```python
reaction_time_raw_df
```

#### `mean_reaction_time`

```python
mean_reaction_time_df = reaction_time_raw_df.groupby('bidder_id').mean().rename({'time':'mean_reaction_time'},axis=1)
```

#### `median_reaction_time`

```python
median_reaction_time_df = reaction_time_raw_df.groupby('bidder_id').median().rename({'time':'median_reaction_time'},axis=1)
```

#### `var_reaction_time`

```python
var_reaction_time_df = reaction_time_raw_df.groupby('bidder_id').var().rename({'time':'var_reaction_time'},axis=1)
```

### reaction_time_df

```python
reaction_time_df = pd.concat([mean_reaction_time_df,median_reaction_time_df,var_reaction_time_df],axis=1)
reaction_time_df
```

## Bid counts

#### `total_bids`
Total bid count per unique bidder

```python
# groupby bidder and count number of bids
total_bids_df = bids_df.groupby('bidder_id').count().reset_index()

# keep bidder_id and another random column, rename the random column to feature name
total_bids_df = total_bids_df[['bidder_id','auction']].rename({'auction':'total_bids'},axis=1).set_index('bidder_id')
```

#### `mean_bids_per_day`

```python
# create empty df
bids_per_day_df = pd.DataFrame(columns=['bidder_id','time'])

# create list of data split into 9 days, to iterate through
days = [day1,day2,day3,day4,day5,day6,day7,day8,day9]

for day in days:
    
    # drop 'auction' as it is not needed
    day = day.drop(columns=['auction'],axis=1)
    
    # count number of bids each unique bidder made
    day_df = day.groupby('bidder_id').count().reset_index()
    
    # add bids counts to the dataframe created above
    bids_per_day_df = pd.concat([bids_per_day_df,day_df],ignore_index=True)

# convert time from object to int
bids_per_day_df['time'] = bids_per_day_df['time'].apply(int)

# get the mean bids per day
mean_bids_per_day_df = bids_per_day_df.groupby('bidder_id').mean()
mean_bids_per_day_df = mean_bids_per_day_df.rename({'time':'mean_bids_per_day'},axis=1)
```

#### relating to bids per auction

```python
# shortcut to get bids_per_auction_df
bids_per_auction_df = pd.read_csv('raw_data/bids_per_auction.csv')
```

```python
days = [day1,day2,day3,day4,day5,day6,day7,day8,day9]

# get unique auctions for each of the 9 days
unique_auctions = []
for day in days:
    unique_auctions.append(list(day.auction.unique()))

# create empty df
bids_per_auction_df = pd.DataFrame(columns=['bidder_id','auction'])

# iterate through the 9 days
for i in range(9):
    day = days[i].drop(columns='time',axis=1)
    unique_auctions_list = unique_auctions[i]
    for auction in unique_auctions_list:
        auction_df = day.loc[day.auction == auction]
        bid_count_df = auction_df.groupby('bidder_id').count().reset_index()
        bids_per_auction_df = pd.concat([bids_per_auction_df,bid_count_df],ignore_index=True)
        
# change values in auction column to int
bids_per_auction_df['auction'] = bids_per_auction_df['auction'].apply(int)
```

##### `mean_bids_per_auction`

```python
mean_bids_per_auction_df = bids_per_auction_df.groupby('bidder_id').mean().rename({'auction':'mean_bids_per_auction'},axis=1)
```

##### `median_bids_per_auction`

```python
median_bids_per_auction_df = bids_per_auction_df.groupby('bidder_id').median().rename({'auction':'median_bids_per_auction'},axis=1)
```

### bids_counts_df

```python
bids_counts_df = pd.concat([mean_bids_per_auction_df,median_bids_per_auction_df,mean_bids_per_day_df,total_bids_df],axis=1)
bids_counts_df
```

#### `mean_time_variance_per_day`

```python
time_variance_per_day_df = pd.DataFrame(columns=['bidder_id','time'])

days = [day1,day2,day3,day4,day5,day6,day7,day8,day9]

for day in days:
    day_df = day.groupby('bidder_id').var().reset_index()
    time_variance_per_day_df = pd.concat([time_variance_per_day_df,day_df],ignore_index=True)

mean_time_variance_per_day_df = time_variance_per_day_df.groupby('bidder_id').mean()
mean_time_variance_per_day_df = mean_time_variance_per_day_df.rename({'time':'mean_time_variance_per_day'},axis=1)
```

#### `first_thirtymins_df`
- bids placed in the first 30mins of the day

```python
one_min = 3157885304.6594987
thirty_mins = one_min*30

days = [day1,day2,day3,day4,day5,day6,day7,day8,day9]

# get start timings
one_day = 4547354838709.678
start = [9631916842105263, 9695580000000000, 9759243157894736]
start_times = []
for i in range(len(start)):
    for _ in range(3):
        start_times.append(start[i])
        start[i]+=one_day


first_thirtymins_df = pd.DataFrame(columns=['bidder_id','time'])

for i in range(9):
    day = days[i]
    day = day.drop(columns=['auction'],axis=1)
    thirtymins = day.loc[(day.time >= start_times[i]) & (day.time < (start_times[i]+thirty_mins))]
    first_thirtymins_df = pd.concat([first_thirtymins_df,thirtymins],ignore_index=True)
    #print(f'Day {i+1} done.') # to track output
    
first_thirtymins_df = first_thirtymins_df.groupby('bidder_id').count().rename({'time':'bids_in_firstthirty'},axis=1)
```

#### `second_thirtymins_df`
- bids placed in the second 30mins of the day

```python
one_min = 3157885304.6594987
thirty_mins = one_min*30

days = [day1,day2,day3,day4,day5,day6,day7,day8,day9]

# get start timings
one_day = 4547354838709.678
start = [9631916842105263, 9695580000000000, 9759243157894736]
start_times = []
for i in range(len(start)):
    for _ in range(3):
        start_times.append(start[i])
        start[i]+=one_day

second_thirtymins_df = pd.DataFrame(columns=['bidder_id','time'])

for i in range(9):
    day = days[i]
    day = day.drop(columns=['auction'],axis=1)
    thirtymins = day.loc[(day.time >= start_times[i]+thirty_mins) & (day.time < (start_times[i]+thirty_mins*2))]
    second_thirtymins_df = pd.concat([second_thirtymins_df,thirtymins],ignore_index=True)
    #print(f'Day {i+1} done.')

second_thirtymins_df = second_thirtymins_df.groupby('bidder_id').count().rename({'time':'bids_in_secondthirty'},axis=1)
```

#### `last_thirtymins_df`
- bids placed in the last 30mins of the day

```python
# bids placed in last 30mins of the day
days = [day1,day2,day3,day4,day5,day6,day7,day8,day9]
one_min = 3157885304.6594987
thirty_mins = one_min*30

# get end timings
one_day = 4547354838709.678
ends = [9631916842105263, 9695580000000000, 9759243157894736]
end_times = []
for i in range(len(ends)):
    for _ in range(3):
        end_times.append(ends[i]+one_day)
        ends[i]+=one_day

last_thirtymins_df = pd.DataFrame(columns=['bidder_id','time'])

for i in range(9):
    day = days[i]
    day = day.drop(columns=['auction'],axis=1)
    thirtymins = day.loc[(day.time > (end_times[i]-thirty_mins)) & (day.time <= end_times[i])]
    last_thirtymins_df = pd.concat([last_thirtymins_df,thirtymins],ignore_index=True)
    #print(f'Day {i+1} done.') # to track output
    
last_thirtymins_df = last_thirtymins_df.groupby('bidder_id').count().rename({'time':'bids_in_lastthirty'},axis=1)
```

#### `secondlast_thirtymins_df`
- bids placed in the second last 30mins of the day

```python
one_min = 3157885304.6594987
thirty_mins = one_min*30

days = [day1,day2,day3,day4,day5,day6,day7,day8,day9]

# get end timings
one_day = 4547354838709.678
ends = [9631916842105263, 9695580000000000, 9759243157894736]
end_times = []
for i in range(len(ends)):
    for _ in range(3):
        end_times.append(ends[i]+one_day)
        ends[i]+=one_day
    
secondlast_thirtymins_df = pd.DataFrame(columns=['bidder_id','time'])

for i in range(9):
    day = days[i]
    day = day.drop(columns=['auction'],axis=1)
    thirtymins = day.loc[(day.time > end_times[i]-(thirty_mins*2)) & (day.time <= (end_times[i]-thirty_mins))]
    secondlast_thirtymins_df = pd.concat([secondlast_thirtymins_df,thirtymins],ignore_index=True)
    #print(f'Day {i+1} done.') # to track output
    
secondlast_thirtymins_df = secondlast_thirtymins_df.groupby('bidder_id').count().rename({'time':'bids_in_secondlastthirty'},axis=1)
```

#### `meanbids_in_20mins_df`
- mean number of bids placed in each 20min interval

```python
one_min = 3157885304.6594987
twenty_df = bids_df[['bidder_id','time']]

start = 9631916842105263 #inclusive
end = start+(20*one_min)

meanbids_in_20mins_df = pd.DataFrame(columns=['bidder_id','time'])

for i in range(2232):
    
    bids = twenty_df.loc[(twenty_df.time >= start) & (twenty_df.time < end)]
    count = bids.groupby('bidder_id').count().reset_index()
    meanbids_in_20mins_df = pd.concat([meanbids_in_20mins_df,count],ignore_index=True)
    
    # update start and end time
    start += (20*one_min)
    end += (20*one_min)
    
    # print(i,end=' ') # track output

meanbids_in_20mins_df['time'] = meanbids_in_20mins_df['time'].apply(int)
meanbids_in_20mins_df = meanbids_in_20mins_df.groupby('bidder_id').mean().rename({'time':'avg_bids_in_20mins'},axis=1)
```

## Bidder Downtime
- Bidding downtime would be the time between a bidder's nth bid and the same bidder's (n+1)th bid

First we extract the time between the bids for each transaction by a bidder

```python
# shortcut to get bidder_downtime_raw_df
bidder_downtime_raw_df = pd.read_csv('raw_data/bidder_downtime_raw.csv')
```

```python
days = [day1,day2,day3,day4,day5,day6,day7,day8,day9]

bidder_downtime_raw_df = pd.DataFrame(columns=['bidder_id','time'])

for day in days:
    
    day = day.drop(columns='auction',axis=1)
    unique_bidders = list(day.bidder_id.unique())
    
    for bidder in unique_bidders:
        bidder_df = day.loc[day.bidder_id == bidder].set_index('bidder_id')
        bidder_df = bidder_df.diff()
        bidder_df = bidder_df.reset_index()
        bidder_downtime_raw_df = pd.concat([bidder_downtime_raw_df,bidder_df],ignore_index=True)
        
    # print('-1 Day Done-') # to track output
```

#### `mean_downtime_df`

```python
mean_downtime_df = bidder_downtime_raw_df.groupby('bidder_id').mean().rename({'time':'mean_bidder_downtime'},axis=1)
```

#### `median_downtime_df`

```python
median_downtime_df = bidder_downtime_raw_df.groupby('bidder_id').median().rename({'time':'median_bidder_downtime'},axis=1)
```

#### `var_downtime_df`

```python
var_downtime_df = bidder_downtime_raw_df.groupby('bidder_id').var().rename({'time':'var_bidder_downtime'},axis=1)
```

### bidder_downtime_df

```python
bidder_downtime_df = pd.concat([mean_downtime_df,median_downtime_df,var_downtime_df],axis=1)
bidder_downtime_df
```

## Old and unclassified features

### bids_per_url_df

```python
bids_per_url_df = bids_df.groupby(['bidder_id','url']).count().reset_index().drop(columns=['merchandise','device','time','country','ip'])
bids_per_url_df = bids_per_url_df.rename({'auction':'bids_per_url'},axis=1)
bids_per_url_df = bids_per_url_df.groupby('bidder_id').aggregate([min, max,'mean','median','count']).bids_per_url
bids_per_url_df = bids_per_url_df.rename(columns={'min': 'min_bids_per_url',
                                                  'max': 'max_bids_per_url',
                                                  'mean': 'mean_bids_per_url',
                                                  'median': 'median_bids_per_url',
                                                  'count': 'num_of_url_used'})
bids_per_url_df
```

### bids_per_ip_df

```python
bids_per_ip_df = bids_df.groupby(['bidder_id','ip']).count().reset_index().drop(columns=['merchandise','device','time','country','url'])
bids_per_ip_df = bids_per_ip_df.rename({'auction':'bids_per_ip'},axis=1)
bids_per_ip_df = bids_per_ip_df.groupby('bidder_id').aggregate([min, max,'mean','median','count']).bids_per_ip
bids_per_ip_df = bids_per_ip_df.rename(columns={'min': 'min_bids_per_ip',
                                                'max': 'max_bids_per_ip',
                                                'mean': 'mean_bids_per_ip',
                                                'median': 'median_bids_per_ip',
                                                'count': 'num_of_ip_used'})
bids_per_ip_df
```

### sets_df

```python
bids_df['bin'] = pd.cut(bids_df['time'], 
                        bins=[0,9660000000000000,9730000000000000,9773000000000000], 
                        labels=["set1", "set2", "set3"])

set1_bids_df = bids_df.loc[bids_df.bin == 'set1'] # transaction in set 1
set2_bids_df = bids_df.loc[bids_df.bin == 'set2'] # transaction in set 2
set3_bids_df = bids_df.loc[bids_df.bin == 'set3'] # transaction in set 3

set1_df = set1_bids_df.groupby('bidder_id').count()
set1_df = set1_df.drop(columns=['device','time','country','ip','url','merchandise','bin']).rename({'auction':'bids_in_set1'},axis=1)

set2_df = set2_bids_df.groupby('bidder_id').count()
set2_df = set2_df.drop(columns=['device','time','country','ip','url','merchandise','bin']).rename({'auction':'bids_in_set2'},axis=1)

set3_df = set3_bids_df.groupby('bidder_id').count()
set3_df = set3_df.drop(columns=['device','time','country','ip','url','merchandise','bin']).rename({'auction':'bids_in_set3'},axis=1)

sets_df = pd.concat([set1_df,set2_df,set3_df],axis=1)

if 'bin' in bids_df.columns:
    bids_df = bids_df.drop(columns=['bin'])
    
sets_df
```

## Inspiried by fakeplastictress
- [ ] The proportion of each bidder’s bids that were in each day ( I believe there were 9 days of bid data).
- [ ] The proportion of each bidder’s bids that were in each of 8 3-hour time slots.
- [ ] The mean time of day for the bids of each bidder.
- [ ] The frequency of each final part of the IP address averaged for each bidder – the first three parts had no positive effect.
- [ ] The average proportion of bots associated with each device code. NB – (difficult to put into CV without data leakage – a slightly different data set for each fold!)
- [ ] Same as above for each country code.
- [X] The total number of bids made in each auction averaged for each bidder. (i.e. did the bidder tend to bid in popular auctions or less popular ones)
- [ ] The maximum number of auctions a bidder has participated in, in one hour.

### auction_bid_count_df
- This is indicative of whether the bidder tend to participate in more popular or less popular auctions

```python
# find total number of bid in each auction
auction_bid_count_df = bids_df.groupby('auction').count()
auction_bid_count_df = auction_bid_count_df[['ip']].rename({'ip':'mean_auction_bid_count_per_bidder'},axis=1).reset_index()

# get unique auctions for unique bidders
unique_auctions_per_bidder = bids_df.groupby(['bidder_id','auction']).count().reset_index()
unique_auctions_per_bidder = unique_auctions_per_bidder[['bidder_id','auction']]

# merge the two df above
auction_bid_count_per_bidder_df = unique_auctions_per_bidder.merge(auction_bid_count_df,how='left',on='auction')
auction_bid_count_per_bidder_df = auction_bid_count_per_bidder_df.drop(columns='auction',axis=1)
mean_auction_bid_count_per_bidder_df = auction_bid_count_per_bidder_df.groupby('bidder_id').mean()

mean_auction_bid_count_per_bidder_df
```

```python
# Feature Set 1: Reaction Time
bids_df_combined['mean_reaction_time'].fillna(bids_df_combined['mean_reaction_time'].median(), inplace = True)
bids_df_combined['median_reaction_time'].fillna(bids_df_combined['median_reaction_time'].median(), inplace = True)
bids_df_combined['var_reaction_time'].fillna(bids_df_combined['var_reaction_time'].median(), inplace = True)

# Feature Set 2: Bids Counts
bids_df_combined['mean_bids_per_auction'].fillna(bids_df_combined['mean_bids_per_auction'].median(), inplace = True)
bids_df_combined['median_bids_per_auction'].fillna(bids_df_combined['median_bids_per_auction'].median(), inplace = True)
bids_df_combined['mean_bids_per_day'].fillna(bids_df_combined['mean_bids_per_day'].median(), inplace = True)
bids_df_combined['total_bids'].fillna(bids_df_combined['total_bids'].median(), inplace = True)

# Feature Set 3: Bidding Activity
bids_df_combined['bids_in_firstthirty'] = bids_df_combined['bids_in_firstthirty'].fillna(0)
bids_df_combined['bids_in_secondthirty'] = bids_df_combined['bids_in_secondthirty'].fillna(0)
bids_df_combined['bids_in_secondlastthirty'] = bids_df_combined['bids_in_secondlastthirty'].fillna(0)
bids_df_combined['bids_in_lastthirty'] = bids_df_combined['bids_in_lastthirty'].fillna(0)
bids_df_combined['mean_time_variance_per_day'].fillna(bids_df_combined['mean_time_variance_per_day'].median(), inplace = True)
bids_df_combined['avg_bids_in_20mins'] = bids_df_combined['avg_bids_in_20mins'].fillna(0)

# Feature Set 4: Bidder Downtime
bids_df_combined['mean_bidder_downtime'].fillna(bids_df_combined['mean_bidder_downtime'].median(), inplace = True)
bids_df_combined['median_bidder_downtime'].fillna(bids_df_combined['median_bidder_downtime'].median(), inplace = True)
bids_df_combined['var_bidder_downtime'].fillna(bids_df_combined['var_bidder_downtime'].median(), inplace = True)

# Feature Set 5: Others
bids_df_combined['mean_time_between_diff_ips'].fillna(bids_df_combined['mean_time_between_diff_ips'].median(), inplace = True)

# Feature Set 6: Old and unclassified
bids_df_combined['min_bids_per_url'].fillna(bids_df_combined['min_bids_per_url'].median(), inplace = True)
bids_df_combined['max_bids_per_url'].fillna(bids_df_combined['max_bids_per_url'].median(), inplace = True)
bids_df_combined['mean_bids_per_url'].fillna(bids_df_combined['mean_bids_per_url'].median(), inplace = True)
bids_df_combined['median_bids_per_url'].fillna(bids_df_combined['median_bids_per_url'].median(), inplace = True)
bids_df_combined['num_of_url_used'].fillna(bids_df_combined['num_of_url_used'].median(), inplace = True)

bids_df_combined['min_bids_per_ip'].fillna(bids_df_combined['min_bids_per_ip'].median(), inplace = True)
bids_df_combined['max_bids_per_ip'].fillna(bids_df_combined['max_bids_per_ip'].median(), inplace = True)
bids_df_combined['mean_bids_per_ip'].fillna(bids_df_combined['mean_bids_per_ip'].median(), inplace = True)
bids_df_combined['median_bids_per_ip'].fillna(bids_df_combined['median_bids_per_ip'].median(), inplace = True)
bids_df_combined['num_of_ip_used'].fillna(bids_df_combined['num_of_ip_used'].median(), inplace = True)

bids_df_combined['bids_in_set1'] = bids_df_combined['bids_in_set1'].fillna(0)
bids_df_combined['bids_in_set2'] = bids_df_combined['bids_in_set2'].fillna(0)
bids_df_combined['bids_in_set3'] = bids_df_combined['bids_in_set3'].fillna(0)

bids_df_combined['unique_devices_used'].fillna(bids_df_combined['unique_devices_used'].median(), inplace = True)
bids_df_combined['unique_countries_used'].fillna(bids_df_combined['unique_countries_used'].median(), inplace = True)

bids_df_combined['mean_bid_per_country'].fillna(bids_df_combined['mean_bid_per_country'].median(), inplace = True)

bids_df_combined['num_of_unique_auctions'].fillna(bids_df_combined['num_of_unique_auctions'].median(), inplace = True)

# Feature set 7: Inspired by fakeplastictrees
bids_df_combined['mean_auction_bid_count_per_bidder'].fillna(bids_df_combined['mean_auction_bid_count_per_bidder'].median(), inplace = True)

# `outcome` column
bids_df_combined['outcome'] = bids_df_combined['outcome'].fillna('Unknown')
```
